**Project Name**: Bioactivity Prediction on Human Acetylcholinesterase

**Aim/Description:**
The project aims to investigate the bioactivity of chemical compounds on Human Acetylcholinesterase, an enzyme crucial for nerve signal regulation. By analyzing how CHEMBL220 interacts with and inhibits this enzyme, the study seeks to evaluate its potential as a therapeutic agent for neurological conditions such as Alzheimer's disease. Bioactivity data will be used to assess the compound's potency, effectiveness, and mechanism of action.

---

**Human Acetylcholinesterase:** This is an enzyme found in the human nervous system. Its main job is to break down a chemical called acetylcholine, which is important for transmitting signals between nerves and muscles. By breaking down acetylcholine, this enzyme helps regulate nerve function.

**Bioactivity data:** This is information about how various substances (like drugs, molecules, or chemicals) interact with a biological target (in this case, Human Acetylcholinesterase). The data shows things like:

How well a substance can inhibit or block the enzyme (which could be important for treating diseases).
How much of a substance is needed to have an effect on the enzyme (called potency).
The type of interaction, like whether the substance activates or inhibits the enzyme.

---

**CHEMBL220** refers to a specific compound or drug entry in the ChEMBL database. Each compound in ChEMBL is assigned a unique identifier, like CHEMBL220, which provides detailed information about that chemical substance. This includes its structure, biological activity, and how it interacts with various biological targets (such as enzymes, receptors, or proteins).

To give more context, **CHEMBL220** is a ChEMBL entry for Tacrine, a drug that
was previously used to treat symptoms of Alzheimer's disease. Tacrine acts as an inhibitor of Human Acetylcholinesterase (the enzyme that breaks down acetylcholine), thereby increasing the levels of acetylcholine in the brain, which helps improve nerve signaling in Alzheimer's patients.

## Importing libraries

This **new_client** object provides access to the ChEMBL web service's API, which allows us to query and retrieve data from the ChEMBL database.

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

## Search for Target protein

```python
# Target search for Acetylcholinesterase
target = new_client.target
target_query = target.search('acetylcholinesterase')
targets = pd.DataFrame.from_dict(target_query)
targets
```

## Select and retrieve bioactivity data for Human Acetylcholinesterase

I am going to assign the first entry to the **selected_target** variable

```python
selected_target = targets.target_chembl_id[0]
selected_target
```

The following line of code lets us access **activity** data—information about how a chemical or drug works, its effects, or its strength from ChEMBL, and then we can use **activity** to search for these details in the database.

```python
activity = new_client.activity
result = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(result)
```

## Labeling compounds as either being active, inactive or intermediate

The bioactivity data is in the IC50 unit. Compounds having values of less than 1000 nM will be considered to be active while those greater than 10,000 nM will be considered to be inactive. As for those values in between 1,000 and 10,000 nM will be referred to as intermediate.

```python
df4 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data/acetylcholinesterase_02_bioactivity_data_preprocessed.csv')
```

```python
bioactivity_threshold = []
for i in df4.standard_value:
  if i >= 10000:
    bioactivity_threshold.append("inactive")
  elif i <= 1000:
    bioactivity_threshold.append("active")
  else:
    bioactivity_threshold.append("intermediate")
```

```python
bioactivity_class = pd.Series(bioactivity_threshold, name='class')
df5 = pd.concat([df4, bioactivity_class], axis=1)
df5.head(10)
```
